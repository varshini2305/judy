```python
import os
import json
import requests

api_key = os.getenv("OPENAI_API_KEY")  # or paste directly for a one-off test
assert api_key, "OPENAI_API_KEY is not set"

url = "https://api.openai.com/v1/responses"
headers = {
    "Authorization": f"Bearer {api_key}",
    "Content-Type": "application/json",
}
payload = {
    "model": "gpt-5.4-nano",
    "input": 'Return JSON only: {"ok": true}'
}

resp = requests.post(url, headers=headers, json=payload, timeout=60)

print("status:", resp.status_code)
try:
    data = resp.json()
except Exception:
    print(resp.text)
    raise

print(json.dumps(data, indent=2)[:4000])
```

```python
from google import genai
from google.genai.types import HttpOptions

VERTEX_PROJECT = "209722657949"
VERTEX_LOCATION = "us"
TUNED_ENDPOINT = "projects/209722657949/locations/us/endpoints/7173389781471068160"
BASE_MODEL = "gemini-3.5-flash"

PROMPT = """SYSTEM PROMPT:
You are a careful question-answering judge. Return only which answer is better: A or B.

QUESTION:
What is 2+2?

ANSWER A:
4

ANSWER B:
5"""

client = genai.Client(
    vertexai=True,
    project=VERTEX_PROJECT,
    location=VERTEX_LOCATION,
    http_options=HttpOptions(api_version="v1"),
)

print("Vertex client ready")
print("project:", VERTEX_PROJECT)
print("location:", VERTEX_LOCATION)
print("tuned endpoint:", TUNED_ENDPOINT)
```

```python
tuned_response = client.models.generate_content(
    model=TUNED_ENDPOINT,
    contents=PROMPT,
)

print("Tuned response:")
print(tuned_response.text)
```

```python
baseline_response = client.models.generate_content(
    model=BASE_MODEL,
    contents=PROMPT,
)

print("Baseline response:")
print(baseline_response.text)
```

```python
print({
    "tuned_endpoint": TUNED_ENDPOINT,
    "tuned_text": tuned_response.text,
    "baseline_model": BASE_MODEL,
    "baseline_text": baseline_response.text,
})
```
